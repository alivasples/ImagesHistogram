```python
import pandas as pd
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
class Image:
    def __init__(self, folder, name, ext):
        self.imagePath = "images/"+folder+name+'.'+ext;
        self.histCountPath = "histograms/"+folder+name+"_count."+ext;
        self.histPercePath = "histograms/"+folder+name+"_perce."+ext;
        self.img = cv2.imread(self.imagePath,0);
        self.df = pd.DataFrame({
            'pixel': np.arange(256),
            'count': np.histogram(self.img.ravel(),256,[0,256])[0],
            'percentage': np.histogram(self.img.ravel(),256,[0,256])[0]*100/len(self.img.ravel())
        });
        del self.img;
    #def show(self):
    #    cv2.namedWindow('image', cv2.WINDOW_NORMAL);
    #    cv2.imshow('image',self.img);
    #    cv2.waitKey(0);
    #    cv2.destroyAllWindows();
    def calcCountHist(self):
        figure = self.df.plot(x='pixel', y='count', kind='bar', figsize=(100,30)).get_figure();
        figure.savefig(self.histCountPath);
    def calcPerceHist(self):
        figure =  self.df.plot(x='pixel', y='percentage', kind='bar', figsize=(100,30)).get_figure();
        figure.savefig(self.histPercePath)
    
        
```

```python
names = ['apple','carrot','orange','pear','pineapple']
ext = 'jpg'
for idxName in range(4,5):
    name = names[idxName]
    folder = name+'s/'
    for i in range(10):
        item = name + str(i+1)
        image = Image(folder,item,ext);
        print(image.imagePath)
        image.calcCountHist()
        image.calcPerceHist()
        del image
```
